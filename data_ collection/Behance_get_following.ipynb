```python
def get_project_info(user, api_key, page):
    api_url = f"http://www.behance.net/v2/users/{user}/following?client_id={api_key}"
    response = requests.get(api_url, params={'page': page})
    return response
```

```python
def get_response_rate(response):    
    if response.status_code == 200:
        return json.loads(response.content.decode('utf-8'))
    elif response.status_code == 404:
        return 'user_done'
    elif response.status_code == 429:
        return 'too_many_requests'
    elif response.status_code == 500:
        return 'server error'
```

```python
def parse_project_response(a, user_id):
    res=pd.DataFrame()
    for j in a['projects']:
        project_dict={}
        project_dict['user_id']=user_id
        project_dict['project_id']=j['id']
        project_dict['created']=j['created_on']
        project_dict['stats']=j['stats']
        project_dict['owners']=[i['id'] for i in j['owners']]
        res=res.append(project_dict, ignore_index=True)
    return res
```

```python
def parse_project_response_following(a, user_id):
    res=pd.DataFrame()
    for j in a['following']:
        project_dict={}
        project_dict['user_id']=user_id
        project_dict['followed_id']=j['id']
        project_dict['followed_user']=j['username']
        project_dict['created']=j['created_on']
        project_dict['stats']=j['stats']
        project_dict['display_name']=j['display_name']
        project_dict['first_name']=j['first_name']
        project_dict['last_name']=j['last_name']
        project_dict['location']=j['location']
        project_dict['fields']=j['fields']
        res=res.append(project_dict, ignore_index=True)
    return res
```
