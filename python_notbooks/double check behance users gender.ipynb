```python
import os
import pandas as pd
```

```python
data_bh=pd.read_csv(data_bh_p+"behance_filtered_sample_min10_act_max5000_following_sample0.csv")
```

```python
path="/Users/orsi/Dropbox/behance-data/behance_following/"
files=os.listdir(path)
```

```python
u_ids=list(data_bh['user_id'])
```

```python
def check_user(row):
    if row['followed_id'] in u_ids:
        return 1
    else:
        return 0
```

```python
res=pd.DataFrame()
for f in files:
    fs=pd.read_csv(path+f)
    fs['u_in']=fs.apply(check_user,1)
    res=res.append(fs[fs['u_in']==1].drop_duplicates(subset=['followed_id'])).drop_duplicates(subset=['followed_id'])
```

```python
len(res)
```

```python
data_with_user_info=data_bh.set_index('user_id').join(res.set_index('followed_id')[['display_name', 'first_name',
      'followed_user', 'last_name', 'location']])
```
