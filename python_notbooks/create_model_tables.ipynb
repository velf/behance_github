```python
def get_sign_follow_suc(row):
    if row['P>|t|']<0.05:
        return 1
    else:
        return 0
    
def get_sign_surv(row):
    if row['P>|z|']<0.05:
        return 1
    else:
        return 0
```

```python
def organize_output(dfs):
    sims=pd.concat(dfs)
    averages=pd.DataFrame(sims.groupby(['order','Unnamed: 0'])["Coef."].mean()).reset_index()
    if 'P>|t|' in sims.columns:
        sims['Sign.']=sims.apply(get_sign_follow_suc,1)
    else:
        sims['Sign.']=sims.apply(get_sign_surv,1)
        
    signs=pd.DataFrame(sims.groupby(['order','Unnamed: 0'])["Sign."].sum()).reset_index()
    max_=pd.DataFrame(sims.groupby(['order','Unnamed: 0'])["Coef."].max()).reset_index()
    min_=pd.DataFrame(sims.groupby(['order','Unnamed: 0'])["Coef."].min()).reset_index()
    
    all_sim=averages[['Coef.']].join(signs[['Sign.']]).join(max_[["Coef."]].rename(columns={"Coef.":"Max"})).join(min_[["Coef."]].rename(columns={"Coef.":"Min"})).rename(columns={"Coef.":'Avg.'})
    return all_sim
        
    
```
